17. Describe the time complexity of accessing elements in a dictionary.

Answer-
Accessing elements in a dictionary is O(1) (constant time), because it uses a hash table internally.

18. Why are dictionaries considered faster than other data structures?

Answer-
Dictionaries use hashing, which allows direct access to values using keys—much faster than searching through a list.

19. What are the collision issues in hash tables, and how do they affect data retrieval?

Answer-
A collision occurs when two keys have the same hash value. Python handles this using techniques like chaining or open addressing, but it can slightly slow down lookups.
